```python
# OOP|4 inheritance and subclasses
# INHERITANCE| allows us to inherit attribute and method from  parent class
# FOR INSTANCE| when developer and manager class needed to be created from the company and want to inherit from Employee class
class Employee:
    # to keep track increment in the number of Employee
    num_of_emp = 0
    
    # creating raise amount variable in the class
    raise_amt = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
        
        # the increment should be place here
        Employee.num_of_emp += 1
    
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amt) # any of this can be used| instance variable
        self.pay = int(self.pay * Employee.raise_amt) # any of this can be used| class variable

# creating Developer class a subclass that will inherit from  the class Employee
class Developer(Employee):
    # to customize Developer class
    raise_amt = 1.10
    
    # creating __init__  method for developer role from the subclass Developer that will handle more info not related to Employee class
    # instantiating init method that will handle role of developer class
    
    def __init__(self, first, last, pay, prog_lang): # the first, last, pay will be handle by Employee class while prog_lang by Developer class
        # Employee.__init__(self, first, last, pay) this can also be used. however super().__init__ is preferred
        
        super().__init__(first, last, pay) # this will pass the first, last, pay to the Employee init method and let that handle those argument.
        self.prog_lang = prog_lang

# creating manager class a subclass that will inherit from the class Employee
class Manager(Employee):
    def __init__(self, first, last, pay, employees=None):
        super().__init__(first, last, pay)
        if employees is None:
            self.employees = []  # creating an empty employee list for the manager to supervise
        else:
            self.employees = employees
    
    # assigning employee for manger to supervise
    def add_emp(self, emp):
        if emp not in self.employees:
            self.employees.append(emp)
            
    # removing the employee from the manager list
    def remove_emp(self, emp):
        if emp in self.employees:
            self.employees.remove(emp)
            
    # accessing the number of employee form the manager desk
    def print_emp(self):
        for emp in self.employees:
            print("=>: ", emp.fullname())
    
    
# calling instance variables by passing argument into each of employee class
# passing a dev variable to the developer class which will be inherited from the Employee class
dev_1 = Developer('adams', 'labaran', 70000, 'front-end')
dev_2 = Developer('isah', 'labaran', 90000, 'back-end')
dev_3 = Developer('musty', 'labaran', 60000, 'data science')

# dev_1, dev_2 inherited from the Employee class
print(dev_1.pay)
print(dev_1.fullname())
print(dev_1.prog_lang)

dev_1.apply_raise() # this is for Employee increment
print(dev_1.pay)

# creating manager and assigning employee more manager can be created here.
mgr_1 = Manager("sue", "Smith", 12000, [dev_1])
print(mgr_1.email)



# adding more employees to the manager's desk
mgr_1.add_emp(dev_2)

# print number oof employee assign to Mgr_1 desk
mgr_1.print_emp()

# removing employee from mgr_1 desk
# mgr_1.remove_emp(dev_1) 

```

```python
# functions in classes are called methods, and at this moment the employee classhas one method dunder inti.
# dunder init is used to inject all the needed information into the object when it is created and will soon learn about how that works
# METHOD| A method is also a function, but its is bound to a class instance which is an object and may interact with everything in that object.
# method interact with object state.
class Employee:
    def __init__(self, first, last, middle, salary):
        self.first = first
        self.middle = middle  # adding middle name to the Employee class diagram
        self.last = last 
        self.salary = salary
        
    # adding a method that will returns full name
     
    def get_full_name(self):
        return f"{self.first} {self.last}"
    
    
    # the Employee class get a new method that rauses the salary by certain percentage.
    # this method does'nt return a value because it is interacted with object state.
    
    def raise_salary(self, percentage): 
        self.salary = self.salary * (100 + percentage) / 100
        
    # add bonus to the Employee class
    def add_bonus(self, amount):
        self.salary = self.salary + amount
        


e1 = Employee("Vera", "Maria", "Schmidt", 2000)
e2 = Employee("Chuck", "Thomas", "murphy", 2000)
e3 = Employee("Dave", "Franz", "Dreissing", 2000)

print(e1.get_full_name(), e1.salary)
print(e2.get_full_name(), e2.salary)
print(e3.get_full_name(), e3.salary)
print()

# apply salary raise
print("# apply salary raise")

# salary raise by 12%
e3.raise_salary(12) 

# bonus
e3.add_bonus(150) 

print(e3.get_full_name(), e3.salary)
```
